```python
import pandas as pd
from random import randrange, choice
from datetime import timedelta, datetime

def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)
```

# Generate Random Data

```python
number_of_records = 100
number_of_stages = 5
stages = ['stage_'+str(i) for i in range(number_of_stages)]
ts_start = datetime.strptime('1/1/2015 1:30 PM', '%m/%d/%Y %I:%M %p')
ts_end = datetime.now()
```

```python
col_id =        [int(i/number_of_stages)       for i in range(number_of_records)]
col_stage =     [choice(stages)                for i in range(number_of_records)]
col_timestamp = [random_date(ts_start, ts_end) for i in range(number_of_records)]
```

```python
data = pd.DataFrame( {
    'col_id': col_id,
    'col_stage': col_stage,
    'col_timestamp': col_timestamp
    })
```

```python
result = []

for name, group in data.groupby('col_id'):
    group_sorted = group.sort('col_timestamp')
    ts = group_sorted['col_timestamp']
    stage = group_sorted['col_stage']
#     print group
#     print group_sorted
    for ts_prev,ts_next, stage_prev, stage_next in zip(ts[:(len(ts)-1)], ts[1:], stage[:(len(stage)-1)], stage[1:]):
        result.append([name, stage_prev, stage_next, ts_next - ts_prev]) 
final = pd.DataFrame(result, columns=['col_id', 'from', 'to', 'timespent'])
```
